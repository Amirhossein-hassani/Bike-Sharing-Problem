## Parameters of the model

```python
# Parameters
I = range(0, 10) # number of stations
T = range(0, 7) # Days
```

```python
# Create a model
model = Model(sense=MINIMIZE)
```

## Defining desicion variables

```python
# Create decision variables
x = {(i, t): model.add_var(var_type=CONTINUOUS, name=f'x_{i}_{t}') for i in I for t in T}  
z = {(i, j, t): model.add_var(var_type=CONTINUOUS, name=f'z_{i}_{j}_{t}') for i in I for j in I for t in T}  
sp = {(i, t): model.add_var(var_type=CONTINUOUS, name=f'sp_{i}_{t}') for i in I for t in T}  
sn = {(i, t): model.add_var(var_type=CONTINUOUS, name=f'sn_{i}_{t}') for i in I for t in T}  
```

## Defining the objective function


```python
# Define the objective function
model.objective = (xsum(xsum(h[i] * sp[i, t] for i in I) for t in T) +
                   xsum(xsum(pi[i] * sn[i, t] for i in I) for t in T) +
                   xsum(xsum(xsum(c.iloc[i,j] * z[i, j, t] for i in I) for j in I) for t in T)
                  )
```

## Defining constraints

```python
# Add constraints
for t in T:
    #model += xsum(sp[i, t] for i in I) == xsum(x[i, t] for i in I) == Q
    for i in I:
        model += x[i, t] <= w[i]
        model += x[i, t] + xsum(y[t].iloc[k,i] for k in I) - xsum(y[t].iloc[i,j] for j in I) == sp[i, t] - sn[i, t]        
        model += x[i, t] >=0
        model += sp[i, t] >=0
        model += sn[i, t] >=0
for t in range(0,6):
    for i in I:
        break
        model += sp[i, t] - xsum(z[i, j, t] for j in I) + xsum(z[k, i, t] for k in I) == x[i, t+1]
for i in I:
    for j in I:
        for t in T:
            model += z[i, j, t] >= 0
```

```python
# Solve the optimization problem
model.optimize()
```

```python
# Print the results
print("Status:", model.status)
print("Objective Value:", model.objective_value)
print("\nDecision Variables:")
for var in model.vars:
    print(f"{var.name}: {var.x}")
```
